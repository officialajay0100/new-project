## Phase 4: Conditional Statements (if, elif, else)

Concepts covered: if, elif, else, nested conditions.

Now we decide the grade of a student based on marks. A conditional statement
lets the program take different actions depending on a condition.

````python
def calculate_grade(marks):
    # This function takes marks and returns a grade.
    # if checks the first condition. If it is True, that block runs and rest are skipped.
    # elif means "else if" -> checked only if the above if was False.
    # else runs only when none of the above conditions were True.

    if marks >= 90:
        grade = "A+"
    elif marks >= 75:
        grade = "A"
    elif marks >= 60:
        grade = "B"
    elif marks >= 40:
        grade = "C"
    else:
        grade = "Fail"

    return grade   # return sends the result back to whoever called the function


def check_pass_fail(marks):
    # This is a simple if-else, only two outcomes possible.
    if marks >= 40:
        return "Pass"
    else:
        return "Fail"


# Testing our function with different marks values
test_marks = 88
print("Marks:", test_marks, "-> Grade:", calculate_grade(test_marks),
      "-> Result:", check_pass_fail(test_marks))

test_marks2 = 35
print("Marks:", test_marks2, "-> Grade:", calculate_grade(test_marks2),
      "-> Result:", check_pass_fail(test_marks2))
````

Concepts covered: while loop, for loop, break, continue, range().

````python
# ---- WHILE LOOP ----
# A while loop keeps running as long as its condition is True.
# We will use this to build a menu that keeps showing until the user chooses to exit.

counter = 1
print("---- while loop demo ----")
while counter <= 5:
    print("This is loop run number", counter)
    counter = counter + 1   # We must change counter, otherwise the loop never stops (infinite loop)

print("\n---- while loop menu demo (type exit to stop) ----")
while True:                 # True is always True, so this loop runs forever...
    user_choice = input("Type 'exit' to stop the demo, or press Enter to continue: ")
    if user_choice == "exit":
        break                # break immediately stops the loop, no matter what
    else:
        print("Continuing the loop...")
        continue             # continue skips the rest of the code and goes to the next loop run

# ---- FOR LOOP ----
# A for loop is used when we want to go through items one by one, a fixed number of times.
# range(5) gives numbers 0, 1, 2, 3, 4 (starts at 0, stops before 5)

print("\n---- for loop demo ----")
for i in range(5):
    print("Processing student number", i + 1)

# for loop is perfect for going through a list of students, which we will do in Phase 7.
````

## Phase 11: Functions

Concepts covered: defining functions, parameters, return values, default parameters,
local vs global variables.

We already used functions earlier, now we understand them properly. A function
lets us write a piece of code once and reuse it many times, which keeps our
project organized.

````python
# def is the keyword used to create a function.
# marks_list is a PARAMETER -> a variable that receives a value when the function is called.

def calculate_total(marks_list):
    total = 0
    for m in marks_list:
        total = total + m
    return total   # return sends the answer back to the code that called this function


def calculate_average(marks_list):
    total = calculate_total(marks_list)   # one function can call another function
    average = total / len(marks_list)
    return average


def calculate_percentage(marks_list, max_marks_per_subject=100):
    # max_marks_per_subject has a DEFAULT VALUE of 100.
    # If the caller does not provide this value, 100 will be used automatically.
    total = calculate_total(marks_list)
    full_marks = max_marks_per_subject * len(marks_list)
    percentage = (total / full_marks) * 100
    return percentage


rahul_marks = [85, 90, 78]

print("Total:", calculate_total(rahul_marks))
print("Average:", calculate_average(rahul_marks))
print("Percentage:", calculate_percentage(rahul_marks))          # uses default 100
print("Percentage (out of 50 each):", calculate_percentage(rahul_marks, 50))  # overrides default

# ---- LOCAL vs GLOBAL VARIABLES ----
college_name = "Delhi Public School"   # this is a GLOBAL variable, available everywhere in the file

def show_college():
    # college_name here refers to the global variable, since we are only reading it
    print("\nCollege Name:", college_name)

def wrong_attempt():
    college_name = "Some Other School"   # this creates a NEW local variable, only inside this function
    print("Inside function:", college_name)

show_college()
wrong_attempt()
show_college()   # global value is unchanged, proves local and global are different
````

## Phase 13: Lambda Functions and List Comprehensions

Concepts covered: lambda functions, list comprehension, sorted() with a key.

A lambda function is a short, one-line function without a name, used when
we need a small function only for a short time. A list comprehension is a
short way to create a list using a single line of code.

````python
students = [
    {"name": "Rahul", "marks": 85},
    {"name": "Priya", "marks": 92},
    {"name": "Aman", "marks": 76},
]

# ---- LAMBDA FUNCTION ----
# lambda argument: expression
# This lambda takes one student dictionary and returns its marks.
get_marks = lambda student: student["marks"]
print("Marks of first student using lambda:", get_marks(students[0]))

# sorted() can take a "key" function to decide HOW to sort.
# Here we sort students by marks, highest first.
sorted_students = sorted(students, key=lambda s: s["marks"], reverse=True)
print("\nStudents sorted by marks (highest first):")
for s in sorted_students:
    print(s["name"], "-", s["marks"])

# ---- LIST COMPREHENSION ----
# Normal way to get a list of only names:
names_normal = []
for s in students:
    names_normal.append(s["name"])
print("\nNames (normal for loop):", names_normal)

# Same result using list comprehension -> shorter and considered more "Pythonic"
names_short = [s["name"] for s in students]
print("Names (list comprehension):", names_short)

# List comprehension with a condition: only students who passed (marks >= 40)
passed_names = [s["name"] for s in students if s["marks"] >= 40]
print("Passed students:", passed_names)

# List comprehension to calculate average using sum() and len() built in functions
all_marks = [s["marks"] for s in students]
average_marks = sum(all_marks) / len(all_marks)
print("\nClass average:", average_marks)
````

## Phase 14: Exception Handling

Concepts covered: try, except, else, finally, raising a custom exception.

When something unexpected happens (like the user typing text instead of a
number), our program should not crash. Exception handling lets us catch the
problem and respond nicely instead of crashing.

````python
# ---- BASIC try-except ----
def get_marks_safely():
    try:
        # The code that might cause an error goes inside try
        marks = float(input("Enter marks: "))
        return marks
    except ValueError:
        # This block runs ONLY if a ValueError happens (wrong type of input)
        print("That was not a valid number. Please enter digits only, like 85 or 85.5")
        return None
    else:
        # else runs only if NO error happened in the try block
        print("Marks entered successfully")
    finally:
        # finally ALWAYS runs, whether there was an error or not
        print("Marks input attempt finished")


result = get_marks_safely()
print("Result:", result)


# ---- CUSTOM EXCEPTION ----
# We can create our own exception type by inheriting from the built in Exception class.
# We will understand "inheritance" fully in Phase 18, for now just know that
# this line means "InvalidMarksError is a special kind of Exception".

class InvalidMarksError(Exception):
    pass   # pass means "do nothing", this class does not need any extra code


def validate_marks(marks):
    if marks < 0 or marks > 100:
        # raise is used to create an error on purpose, when our own rule is broken
        raise InvalidMarksError(f"Marks must be between 0 and 100, but got {marks}")
    return marks


try:
    validate_marks(150)   # this will break our rule on purpose, to show the error
except InvalidMarksError as error:
    print("\nCustom Error Caught:", error)

# This is how our final project will make sure no wrong marks ever get saved.
````

## Phase 15: File Handling

Concepts covered: open(), write mode, read mode, append mode, with statement, CSV file basics.

Till now, all our data disappeared when the program stopped. File handling
lets us save data permanently on the disk, and read it back later.

````python
import csv   # csv module helps us work with CSV files, which act like simple spreadsheets

# ---- WRITING TO A FILE ----
# "with" automatically closes the file for us once we are done, which is the safe way to work with files.
# "w" mode means write mode. It creates a new file, or erases an old file with the same name.

with open("students.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Roll Number", "Marks"])   # header row
    writer.writerow(["Rahul", 101, 85])
    writer.writerow(["Priya", 102, 92])
    writer.writerow(["Aman", 103, 76])

print("Data written to students.csv successfully")

# ---- APPENDING TO A FILE ----
# "a" mode means append mode. It adds new lines WITHOUT erasing the old data.
with open("students.csv", "a", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Sneha", 104, 88])

print("New student appended to students.csv")

# ---- READING FROM A FILE ----
# "r" mode means read mode. This is used only to read existing data.
print("\n---- Reading students.csv ----")
with open("students.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)

# This is exactly how our final project will save data so it is not lost
# when we close the program.
````

## Phase 17: OOP - Classes and Objects

Concepts covered: class, object, __init__ constructor, self, instance methods, instance attributes.

Until now we stored one student as a dictionary. In real projects, we use
CLASSES to bundle data and the actions on that data together. A class is
like a blueprint, and an object is a real item built from that blueprint.

````python
class Student:
    # __init__ is a special method called the CONSTRUCTOR.
    # It runs automatically the moment a new Student object is created.
    # self refers to the exact object being created or used. Every method inside
    # a class must have self as its first parameter.

    def __init__(self, name, roll_number, marks):
        self.name = name                 # self.name is an INSTANCE ATTRIBUTE, unique to this object
        self.roll_number = roll_number
        self.marks = marks               # marks will be a list, like [85, 90, 78]

    def total_marks(self):
        # This is an INSTANCE METHOD, a function that belongs to the class
        # and can use self to access this object's own data.
        return sum(self.marks)

    def average_marks(self):
        return self.total_marks() / len(self.marks)

    def get_grade(self):
        avg = self.average_marks()
        if avg >= 90:
            return "A+"
        elif avg >= 75:
            return "A"
        elif avg >= 60:
            return "B"
        elif avg >= 40:
            return "C"
        else:
            return "Fail"

    def display(self):
        # A method to neatly print this student's full report
        print(f"Name: {self.name}")
        print(f"Roll Number: {self.roll_number}")
        print(f"Marks: {self.marks}")
        print(f"Total: {self.total_marks()}")
        print(f"Average: {self.average_marks():.2f}")
        print(f"Grade: {self.get_grade()}")


# Creating OBJECTS from the Student class. Each object is a separate, independent student.
student1 = Student("Rahul", 101, [85, 90, 78])
student2 = Student("Priya", 102, [95, 92, 98])

print("---- Student 1 ----")
student1.display()

print("\n---- Student 2 ----")
student2.display()

# Each object keeps its own data separately, even though they came from the same class.
print("\nStudent 1 name:", student1.name)
print("Student 2 name:", student2.name)
````

## Phase 18: OOP - Inheritance

Concepts covered: inheritance, super(), method overriding, extending a parent class.

Inheritance lets a new class reuse everything from an existing class, and
add or change only what is different. This avoids writing the same code again.

````python
class ScholarshipStudent(Student):
    # Student is the PARENT class (also called base class).
    # ScholarshipStudent is the CHILD class (also called derived class).
    # ScholarshipStudent gets everything Student already has, automatically.

    def __init__(self, name, roll_number, marks, scholarship_amount):
        # super() lets us call the parent class's __init__, so we do not repeat that code.
        super().__init__(name, roll_number, marks)
        self.scholarship_amount = scholarship_amount   # this is a NEW attribute, only for this child class

    def display(self):
        # This is METHOD OVERRIDING: we are writing a new version of display()
        # that replaces the parent's version, but we still reuse the parent's work first.
        super().display()   # this runs the parent class's display() first
        print(f"Scholarship Amount: Rs {self.scholarship_amount}")   # then we add our own extra line


# ScholarshipStudent object has everything Student has, PLUS scholarship_amount
topper = ScholarshipStudent("Aman", 103, [98, 96, 99], 50000)
topper.display()

# It IS a Student too, because of inheritance
print("\nIs topper a Student object?", isinstance(topper, Student))
````

## Phase 19: OOP - Encapsulation and Polymorphism

Concepts covered: private attributes (using __), property/getter idea, polymorphism.

Encapsulation means hiding the internal details of a class and only allowing
controlled access to them. Polymorphism means the same method name behaving
differently for different classes.

````python
class SecureStudent:
    def __init__(self, name, marks):
        self.name = name
        self.__marks = marks   # double underscore before a name makes it PRIVATE
        # Private means it cannot be accessed directly from outside the class easily.
        # This protects important data from being changed by mistake.

    def get_marks(self):
        # This is called a GETTER. It gives safe, read only access to private data.
        return self.__marks

    def set_marks(self, new_marks):
        # This is called a SETTER. It allows changing private data, but with our own rules.
        if 0 <= new_marks <= 100:
            self.__marks = new_marks
        else:
            print("Invalid marks. Marks must be between 0 and 100")


secure_student = SecureStudent("Sneha", 88)
print("Marks using getter:", secure_student.get_marks())

secure_student.set_marks(95)          # valid update, allowed
print("Marks after valid update:", secure_student.get_marks())

secure_student.set_marks(150)         # invalid update, rejected by our own rule
print("Marks after invalid update attempt:", secure_student.get_marks())

# Trying secure_student.__marks directly from outside would not work normally,
# because it is private. Always use get_marks() and set_marks() instead.


# ---- POLYMORPHISM ----
# Polymorphism means "many forms". The same method name display_result()
# behaves differently depending on which class's object is calling it.

class RegularStudent:
    def display_result(self):
        print("Regular Student: Result declared through normal exam.")

class SportsStudent:
    def display_result(self):
        print("Sports Student: Result declared with sports quota grace marks.")

# One single loop calls the SAME method name on different types of objects,
# and each one behaves in its own way. This is polymorphism in action.
all_students = [RegularStudent(), SportsStudent()]
for s in all_students:
    s.display_result()
````

## Phase 20: Iterators and Generators

Concepts covered: iterator, iter(), next(), generator function, yield keyword.

A generator is a special kind of function that produces values one at a
time, only when needed. This saves memory when we have a huge amount of
student data, because we do not keep everything in memory together.

````python
# ---- ITERATOR BASICS ----
# A list is "iterable", meaning we can go through it. iter() turns it into an ITERATOR.
names = ["Rahul", "Priya", "Aman"]
name_iterator = iter(names)

# next() gives one item at a time from the iterator, moving forward each time
print("First call to next():", next(name_iterator))
print("Second call to next():", next(name_iterator))
print("Third call to next():", next(name_iterator))
# A fourth next() call here would give a StopIteration error, because the list has ended.


# ---- GENERATOR FUNCTION ----
# yield is used instead of return. yield gives one value, PAUSES the function,
# and continues from the same place the next time it is asked for another value.

def student_generator(student_list):
    for student in student_list:
        print(f"(generator is now preparing data for {student['name']})")
        yield student   # pause here, hand over one student, wait for the next request


students_data = [
    {"name": "Rahul", "marks": 85},
    {"name": "Priya", "marks": 92},
    {"name": "Aman", "marks": 76},
]

gen = student_generator(students_data)

print("\n---- Pulling students one by one from the generator ----")
print(next(gen))
print(next(gen))
print(next(gen))

# A generator does not calculate everything at once, it works only when asked.
# This is very useful when a school has lakhs of student records.
````

## Phase 21: Decorators

Concepts covered: decorators, functions as arguments, wrapper functions, @ syntax.

A decorator is a function that takes another function, adds some extra
behaviour around it, and returns a new function. We use it here to measure
how long our result calculation takes.

````python
# This is our decorator.
# It receives another function as input.
def my_decorator(func):

    # Wrapper is a new function.
    # It adds extra code before and after the original function.
    def wrapper():

        # Runs before the original function
        print("Before function")

        # Call the original function
        func()

        # Runs after the original function
        print("After function")

    # Return the wrapper function
    return wrapper


# Apply the decorator to the function
@my_decorator
def say_hello():

    # Original function
    print("Hello Students!")


# Call the function
say_hello()
````

# Python Decorators

## What is a Decorator?

A **Decorator** is a function that **adds extra functionality** to another function **without changing the original function's code**.

In simple words,

> **Decorator = Extra Code + Original Function**

The original function remains unchanged. The decorator simply runs some extra code **before** or **after** the original function.

---

# Example

Suppose we have a simple function.

```python
def say_hello():
    print("Hello Students!")
```

Output

```
Hello Students!
```

Now suppose we want to print something **before** and **after** this function without modifying it.

For this, we use a **Decorator**.

---

# Decorator Code

```python
def my_decorator(func):

    def wrapper():

        print("Before function")

        func()

        print("After function")

    return wrapper
```

### Understanding the Code

### Line 1

```python
def my_decorator(func):
```

- `my_decorator` is a normal Python function.
- It receives another function as its input.
- That input function is stored in the variable **func**.

---

### Line 2

```python
def wrapper():
```

- Inside the decorator, we create another function named **wrapper**.
- This wrapper will contain the extra code.

---

### Line 3

```python
print("Before function")
```

This code executes **before** the original function.

---

### Line 4

```python
func()
```

This line calls the **original function**.

Remember,

```
func = original function
```

---

### Line 5

```python
print("After function")
```

This code executes **after** the original function.

---

### Line 6

```python
return wrapper
```

The decorator returns the wrapper function.

Now the wrapper becomes the new version of the original function.

---

# Applying the Decorator

```python
@my_decorator
def say_hello():
    print("Hello Students!")
```

The `@` symbol means

> Apply this decorator to the function below.

Python automatically converts this code into

```python
def say_hello():

say_hello = my_decorator(say_hello)
```

You never write this line yourself.

Python writes it internally.

## Phase 22: Final Integration - Complete Project

Concepts covered: everything from Phase 1 to Phase 21, combined into one
single, working, menu-driven program.

This is the real, final project. It uses classes, file handling, exception
handling, loops, functions, decorators and more, all working together.
Run this cell and use the menu to add, view, and save student records.

````python
import csv
import time

# ---- CUSTOM EXCEPTION (Phase 14) ----
class InvalidMarksError(Exception):
    pass


# ---- DECORATOR (Phase 21) ----
def time_tracker(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"[time taken by '{func.__name__}': {end_time - start_time:.5f} seconds]")
        return result
    return wrapper


# ---- PARENT CLASS (Phase 17) ----
class Student:
    def __init__(self, name, roll_number, marks):
        self.name = name
        self.roll_number = roll_number
        self.__marks = marks              # private attribute (Phase 19 - Encapsulation)

    def get_marks(self):                  # getter (Phase 19)
        return self.__marks

    def set_marks(self, new_marks_list):  # setter with validation (Phase 19 + Phase 14)
        for m in new_marks_list:
            if m < 0 or m > 100:
                raise InvalidMarksError(f"Invalid marks found: {m}. Marks must be 0-100")
        self.__marks = new_marks_list

    def total_marks(self):
        return sum(self.__marks)

    def average_marks(self):
        return self.total_marks() / len(self.__marks)

    def get_grade(self):
        avg = self.average_marks()
        if avg >= 90:
            return "A+"
        elif avg >= 75:
            return "A"
        elif avg >= 60:
            return "B"
        elif avg >= 40:
            return "C"
        else:
            return "Fail"

    def display(self):
        print(f"Roll No: {self.roll_number} | Name: {self.name} | "
              f"Marks: {self.get_marks()} | Average: {self.average_marks():.2f} | "
              f"Grade: {self.get_grade()}")

    def to_row(self):
        # Converts this student object into a simple list, ready to be written to a CSV file
        return [self.name, self.roll_number, self.get_marks(), self.get_grade()]


# ---- CHILD CLASS (Phase 18 - Inheritance) ----
class ScholarshipStudent(Student):
    def __init__(self, name, roll_number, marks, scholarship_amount):
        super().__init__(name, roll_number, marks)
        self.scholarship_amount = scholarship_amount

    def display(self):                    # method overriding (Phase 18) - polymorphism (Phase 19) too
        super().display()
        print(f"    Scholarship: Rs {self.scholarship_amount}")


# ---- GENERATOR (Phase 20) ----
def student_report_generator(students_list):
    # Yields one student's report line at a time, instead of building everything in memory together
    for s in students_list:
        yield f"{s.name} ({s.roll_number}) - Grade {s.get_grade()}"


# ---- MAIN DATA STORE (Phase 7 - list, Phase 10 - set) ----
all_students = []                # list of Student objects
used_roll_numbers = set()        # set makes sure roll numbers never repeat


# ---- FUNCTIONS FOR EACH MENU OPTION (Phase 11, Phase 12) ----
def add_student():
    name = input("Enter student name: ").strip().title()   # string methods from Phase 6
    roll_number = int(input("Enter roll number: "))

    if roll_number in used_roll_numbers:                     # Phase 10 - checking inside a set
        print("This roll number already exists. Student not added.")
        return

    subject_count = int(input("How many subjects? "))
    marks = []
    for i in range(subject_count):                            # for loop from Phase 5
        m = float(input(f"Enter marks for subject {i + 1}: "))
        marks.append(m)                                       # list append from Phase 7

    try:
        is_scholarship = input("Is this a scholarship student? (yes/no): ").lower() == "yes"
        if is_scholarship:
            amount = float(input("Enter scholarship amount: "))
            new_student = ScholarshipStudent(name, roll_number, marks, amount)
        else:
            new_student = Student(name, roll_number, marks)

        new_student.set_marks(marks)     # this can raise InvalidMarksError, handled below
        all_students.append(new_student)
        used_roll_numbers.add(roll_number)
        print(f"Student {name} added successfully.")

    except InvalidMarksError as error:   # exception handling from Phase 14
        print("Could not add student. Reason:", error)


def view_all_students():
    if len(all_students) == 0:
        print("No students added yet.")
        return
    print("\n---- All Students ----")
    for s in all_students:               # calls display(), which behaves differently
        s.display()                      # for Student vs ScholarshipStudent (polymorphism)


def view_reports_using_generator():
    print("\n---- Student Reports (using generator) ----")
    for line in student_report_generator(all_students):
        print(line)


def show_topper():
    if len(all_students) == 0:
        print("No students added yet.")
        return
    # lambda + built in max() function, similar idea to Phase 13
    topper = max(all_students, key=lambda s: s.average_marks())
    print("\nClass Topper:")
    topper.display()


@time_tracker
def save_to_file():
    # File handling from Phase 15
    with open("final_students.csv", "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Name", "Roll Number", "Marks", "Grade"])
        for s in all_students:
            writer.writerow(s.to_row())
    print("All student data saved to final_students.csv")


def load_from_file():
    try:
        with open("final_students.csv", "r") as file:
            reader = csv.reader(file)
            next(reader)     # skip header row, uses next() from Phase 20
            print("\n---- Data Loaded From File ----")
            for row in reader:
                print(row)
    except FileNotFoundError:
        print("No saved file found yet. Please save data first using option 5.")


# ---- MAIN MENU LOOP (Phase 5 - while loop, Phase 4 - if/elif/else) ----
def run_menu():
    while True:
        print("\n========== STUDENT MANAGEMENT MENU ==========")
        print("1. Add a new student")
        print("2. View all students")
        print("3. View reports (generator based)")
        print("4. Show class topper")
        print("5. Save all data to file")
        print("6. Load data from file")
        print("7. Exit")

        choice = input("Enter your choice (1-7): ")

        if choice == "1":
            add_student()
        elif choice == "2":
            view_all_students()
        elif choice == "3":
            view_reports_using_generator()
        elif choice == "4":
            show_topper()
        elif choice == "5":
            save_to_file()
        elif choice == "6":
            load_from_file()
        elif choice == "7":
            print("Thank you for using Smart Student Management System. Goodbye!")
            break             # break stops the while loop, ending the program
        else:
            print("Invalid choice. Please enter a number between 1 and 7.")


# Uncomment the line below and run this cell to actually use the menu.
run_menu()

print("Program loaded successfully.")
print("Remove the # in front of run_menu() above, and run this cell again to start using the system.")
````
