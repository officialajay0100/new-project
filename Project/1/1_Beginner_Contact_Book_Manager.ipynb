# Project 1 (Beginner): Contact Book Manager

**What will you learn in this project?**
- Real use of Dictionary and List
- Writing and calling functions
- File handling (saving/loading data in a JSON file)
- Basic error handling (try-except)

**Think of it like this:**
Imagine you have a **notebook (diary)** where you write down your friends' names and phone numbers. This project is the digital version of that diary — the only difference is it runs in Python, and the data is saved permanently in a JSON file (like closing your diary and picking up right where you left off next time).

**Project Flow:**
1. Add a contact (name, phone, email)
2. View all contacts
3. Search a contact by name
4. Delete a contact
5. Data keeps saving/loading from a file automatically


## Step 1: Import the required libraries
We use the `json` library to save data in a file — it is the easiest way to store a Python dictionary as a file.

````python
import json
import os

FILE_NAME = "contacts.json"
````

## Step 2: Functions to Load and Save data

- `load_contacts()` → loads all saved contacts from the file (returns an empty dictionary if the file doesn't exist yet)
- `save_contacts()` → writes the current contacts into the file

These two functions act as our "backend".

````python
def load_contacts():
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r") as f:
            return json.load(f)
    return {}

def save_contacts(contacts):
    with open(FILE_NAME, "w") as f:
        json.dump(contacts, f, indent=4)
````

## Step 3: Adding a Contact

Each contact will be a dictionary inside a dictionary:
```
{
  "Rahul": {"phone": "9876543210", "email": "rahul@gmail.com"}
}
```
This lets us search directly by name — just like your phone's contact list works.

````python
def add_contact(contacts):
    name = input("Enter name: ").strip()
    phone = input("Enter phone number: ").strip()
    email = input("Enter email: ").strip()

    if name in contacts:
        print(f"⚠️  '{name}' already exists. It will be overwritten.")

    contacts[name] = {"phone": phone, "email": email}
    save_contacts(contacts)
    print(f"✅ Contact '{name}' added successfully!")
````

## Step 4: Viewing All Contacts

````python
def view_contacts(contacts):
    if not contacts:
        print("📭 No contacts found. Add one first.")
        return

    print("\n----- Your Contact List -----")
    for name, info in contacts.items():
        print(f"👤 {name} | 📞 {info['phone']} | ✉️  {info['email']}")
    print("------------------------------\n")
````

## Step 5: Searching a Contact by Name

````python
def search_contact(contacts):
    name = input("Which name do you want to search? ").strip()
    if name in contacts:
        info = contacts[name]
        print(f"✅ Found! 👤 {name} | 📞 {info['phone']} | ✉️  {info['email']}")
    else:
        print(f"❌ No contact found with the name '{name}'.")
````

## Step 6: Deleting a Contact

We use `try-except` here so that if a wrong name is entered, the program doesn't crash — instead it shows a friendly message.

````python
def delete_contact(contacts):
    name = input("Which contact do you want to delete? ").strip()
    try:
        del contacts[name]
        save_contacts(contacts)
        print(f"🗑️  '{name}' deleted successfully.")
    except KeyError:
        print(f"❌ No contact exists with the name '{name}'.")
````

## Step 7: Main Menu — Putting it all together

This function runs a **loop** that keeps asking the user what to do, until they choose "Exit".

> Note: `input()` works fine in Colab — just run the cell and type in the box that appears below it.

````python
def main():
    contacts = load_contacts()

    menu = '''
    ===== CONTACT BOOK MANAGER =====
    1. Add Contact
    2. View All Contacts
    3. Search Contact
    4. Delete Contact
    5. Exit
    ================================
    '''

    while True:
        print(menu)
        choice = input("Enter your choice (1-5): ").strip()

        if choice == "1":
            add_contact(contacts)
        elif choice == "2":
            view_contacts(contacts)
        elif choice == "3":
            search_contact(contacts)
        elif choice == "4":
            delete_contact(contacts)
        elif choice == "5":
            print("👋 Goodbye! Your contacts are safely saved.")
            break
        else:
            print("⚠️  Please choose a valid option (1-5).")

# Uncomment this line to run the interactive menu in Colab:
# main()
````
